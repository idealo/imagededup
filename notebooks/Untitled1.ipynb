```python
# import keras dependencies 
from keras.models import Model
from keras.applications import MobileNet as CNN
from keras.applications.mobilenet import preprocess_input, decode_predictions
from keras.preprocessing.image import ImageDataGenerator

# Get the model (Mobilenet Pretrained on ImageNet dataset)
model = CNN(include_top=True)
model.summary()

model_cutted = Model(inputs=model.input, outputs=model.layers[-3].output)
model_cutted.summary()


img_gen_train = ImageDataGenerator(preprocessing_function=preprocess_input)
TARGET_SIZE = (224, 224)
BATCH_SIZE = 64
IMG_DIR_TRAIN = '/Users/tanuj.jain/Documents/dedup-data/base_images_copy/'


img_batches_train = img_gen_train.flow_from_directory(
    directory=IMG_DIR_TRAIN,
    target_size=TARGET_SIZE,
    batch_size=BATCH_SIZE,
    color_mode='rgb',
    shuffle=False
)

feat_vecs_train = model_cutted.predict_generator(img_batches_train, len(img_batches_train), verbose=1)
```
